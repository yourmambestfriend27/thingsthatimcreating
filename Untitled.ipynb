```python
def GiveWorkingDay():
    from datetime import date
    from datetime import timedelta
    
    day = date.today()
    if day.weekday() == 5:
        workingday = day + timedelta(days=2)
    elif day.weekday() == 6:
        workingday = day + timedelta(days=1)
    else:
        workingday = day
    print('Working day for',day,'is',workingday)
    
    return
```

```python
#ile dni do konca roku 
def daystoendYear():
    from datetime import date
    date_today = date.today()
    current_year = date_today.year
    date_end_year = date(current_year,12,31)
    delta = date_end_year - date_today
    print(delta.days, 'days to end of the year.')
    return
```

```python
def GiveWorkingDay(year,month,day):
    from datetime import date
    from datetime import timedelta
    
    day = date(year,month,day)
    if day.weekday() == 5:
        workingday = day + timedelta(days=2)
    elif day.weekday() == 6:
        workingday = day + timedelta(days=1)
    else:
        workingday = day
    print('Working day for',day,'is',workingday)
    
    return
```

```python
GiveWorkingDay(2010,9,29)
GiveWorkingDay(2010,9,19)
GiveWorkingDay(2010,10,19)
GiveWorkingDay(2010,5,9)
GiveWorkingDay(year=2020,month=9,day=4)
GiveWorkingDay(day=4,year=2020,month=4)
```

```python
#zadanie 1. 
def PrintAnimal(animal):
        Print_Cat = r'''
    |\---/|
    | o_o |
     \_^_/'''

        Print_Bear = r'''
    /  \.-"""-./  \
    \    -   -    /
     |   o   o   |
     \  .-'"'-.  /
      '-\__Y__/-'   '''

        Print_Bat = r'''
       /\                 /\
      / \'._   (\_/)   _.'/ \
     /_.''._'--('.')--'_.''._\
     | \_ / `;=/ " \=;` \ _/ |
      \/ `\__|`\___/`|__/`  \/
              \(/|\)/       '''

        if animal == 'cat':
            print(Print_Cat)
        elif animal == 'bear':
            print(Print_Bear)
        elif animal == 'bat':
            print(Print_Bat)
        else:
            print('Cannot print %s.\nCorrect values for the parameter are:cat,bear,bat.' %(animal))
        
        return
```

```python
PrintAnimal('cat')
PrintAnimal('bat')
PrintAnimal('bear')
print('\n')
PrintAnimal('rat')
```

```python
#ile dni do konca roku 
def daystoendYear(year,month,day):
    from datetime import date
    date_today = date(year,month,day)
    date_end_year = date(year,12,31)
    delta = date_end_year - date_today
    print(delta.days, 'days to end of the year.')
    return
```

```python
daystoendYear(2020,12,20)
daystoendYear(2020,12,1)
daystoendYear(year=2020,month=1,day=12)
daystoendYear(day=12,year=1992,month=9)
```
